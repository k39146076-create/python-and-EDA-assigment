```python
#Q5. count vowels

def cnt_vowel(s1):
    i=0
    cnt=0
    n=len(s1)
    for i in range (n):

       if (s1[i]=='a' or s1[i]=='e' or s1[i]=='o' or s1[i]=='i' or s1[i]=="u"):
        cnt+=1
    return cnt        
s2="elephant" 
cnt_vowel(s2)    
```
